```python
from nilearn import plotting, image
import nibabel as nib
import os
import numpy as np
from pathlib import Path
import tarfile
```

```python
path_file = 'HCP_PTN1200_recon2/groupICA_3T_HCP1200_MSMAll.tar.gz'
extract_dir = 'data'
with tarfile.open(path_file, 'r:gz') as tar:
    tar.extractall(path = extract_dir)
    
```

```python
for root, dirs, files in os.walk(extract_dir):
    for file in files:
        print(f'Found file: {file}')
```

```python
nii_file = Path(extract_dir) / 'groupICA/groupICA_3T_HCP1200_MSMAll_d15.ica/melodic_IC_sum.nii.gz'
```

```python
img = nib.load(nii_file)
data = img.get_fdata()
```

```python
first_rsn = image.index_img(img, 0)
```

```python
plotting.plot_stat_map(first_rsn)
plotting.view_img(first_rsn)
plotting.show()
```

```python
count = 0
for i in image.iter_img(img):
    if count > 10:
        break
    plotting.plot_stat_map(
        i, display_mode = 'x', cut_coords=1
    )
    count += 1
```
